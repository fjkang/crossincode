```python
import tushare as ts
import pandas as pd 
import numpy as np
import pyecharts
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB

from pyecharts.faker import Faker
import pyecharts.options as opts
from pyecharts.charts import Liquid, Gauge, WordCloud, Line
```

1.导入 tushare 数据包，提取 A 股基本数据信息，取毛利率 gpr 和净利率 npr 的平均数，并分别以水球图和仪 表盘图表现出来(阅读参考文章 2、3);

```python
df = ts.get_stock_basics('20200121')
df
```

```python
gpr = df['gpr'].mean()
npr = df['npr'].mean()
```

```python
lq = Liquid()
lq.add('A股毛利率', [gpr/100])
lq.set_global_opts(title_opts=opts.TitleOpts(title="A股毛利率"))
lq.load_javascript()
```

```python
lq.render_notebook()
```

```python
gauge = Gauge()
gauge.add('', [('A股净利率',round(npr,2))])
gauge.load_javascript()
```

```python
gauge.render_notebook()
```

2.统计 A 股各行业公司数量，并跟进行业分布数据作词云图(阅读参考文章 4、5);

```python
industry_company_count = df.groupby('industry')['name'].agg([('行业公司数量', 'count')])
industry_company_count
```

```python
industry = industry_company_count.index.values.tolist()
count = industry_company_count['行业公司数量'].values.tolist()
```

```python
wc = WordCloud()
wc.add("", zip(industry, count))
wc.set_global_opts(title_opts=opts.TitleOpts(title="各行业公司数量"))
wc.load_javascript()
```

```python
wc.render_notebook()
```

3.利用 tushare 提取茅台数据中的最高最低价，作双折线图，茅台股票代码 600519 (提示:提取单个股票信息 df = ts.get_hist_data('600519',start=' ',end=' ') )(阅读参考文章 1);

```python
df3 = ts.get_hist_data('600519')
df3
```

```python
maotai_h_l_data = df3[['high', 'low']]
line_x = maotai_h_l_data.index.values.tolist()
line_y_h = maotai_h_l_data['high'].values.tolist()
line_y_l = maotai_h_l_data['low'].values.tolist()
```

```python
line = Line()
line.add_xaxis(line_x)
line.add_yaxis('high', line_y_h)
line.add_yaxis('high', line_y_l)
line.set_global_opts(title_opts=opts.TitleOpts(title="茅台最高最低股价趋势"))
line.load_javascript()
```

```python
line.render_notebook()
```
