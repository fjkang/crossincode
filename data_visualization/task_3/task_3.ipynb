1.读取 CSV 格式文件“task_3.1_zhihu_timeline_answer.csv”，将数据命名为 zhihu_data_answer，并查看前部数据(阅读参考文章 1、2);

```python
f = open('./task_3.1_zhihu_timeline_answer.csv', 'rb')
zhihu_data_answer = pd.read_csv(f)
f.close()
zhihu_data_answer.head()
```

2.提取“task_3.1_zhihu_timeline_answer.csv”中【回答点赞数】和【回答感谢数】列数据，制成散点图， 观察两者的关系(阅读参考文章 3、4);

```python
plt_1_answer_like = zhihu_data_answer['回答点赞数'].tolist()
plt_1_answer_thx = zhihu_data_answer['回答感谢数'].tolist()
```
