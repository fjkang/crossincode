7.利用 seed 生成一组固定的随机数 np.random.seed(1)，并用此组成模拟的资金价值曲线

values = np.random.randn(1000).cumsum()

请利用 matplotlib 作出该资金价值曲线图

```python
np.random.seed(1)
values = np.random.randn(1000).cumsum()
line = plt.plot(values, label="模拟的资金价值曲线")
plt.legend()
plt.show()
```
